```python
import pandas as pd
```

```python
df = pd.read_csv('/content/datset_hr.csv')
```

#data split

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['Attrition'],axis = 1)

y = df['Attrition']

X.columns
```

```python
from imblearn.over_sampling import SMOTE, RandomOverSampler
```

```python
smote= SMOTE()
X,y=smote.fit_resample(X,y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 107)
```

#svm

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

from sklearn import svm
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
svm=SVC()
prob= svm.fit(X_train, y_train)
y_pred_svm= prob.predict(X_test)
```

```python
print("Accuracy Linear: ", accuracy_score(y_test, y_pred_svm)* 100, '%')
print("\n")
print('Confusion Matrix:')
print(pd.DataFrame(confusion_matrix(y_test, y_pred_svm)))
print('------------------------------------------------')
print('Classification Report:')
print(classification_report(y_test, y_pred_svm))
```

#xgbClassifier

```python
from xgboost import XGBClassifier
```

```python
xgb_model = XGBClassifier()
xgb_model.fit(X_train, y_train)
y_pred = xgb_model.predict(X_test)

print(classification_report(y_test, y_pred))
```

#naive bayes

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, confusion_matrix
CNB = GaussianNB()
CNB.fit(X_train, y_train)

from sklearn import metrics
predicted = CNB.predict(X_test)
# print(predicted)
print(X_test)
accuracy_score = metrics.accuracy_score(predicted, y_test)
```

```python
print('GussianNB model accuracy is',str('{:04.2f}'.format(accuracy_score*100))+'%')
print('------------------------------------------------')
print('Confusion Matrix:')
print(pd.DataFrame(confusion_matrix(y_test, predicted)))
print('------------------------------------------------')
print('Classification Report:')
print(classification_report(y_test, predicted))
```
